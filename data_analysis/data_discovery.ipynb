```
from pyspark.sql.functions import col,countDistinct
```

# Read

```
path="dbfs:/FileStore/shared_uploads/dtuworkshop@outlook.com"
sales_target = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(f"{path}/Sales_target.csv")
sales_order_header = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("dbfs:/FileStore/shared_uploads/dtuworkshop@outlook.com/List_of_Orders.csv")
sales_order_detail = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("dbfs:/FileStore/shared_uploads/dtuworkshop@outlook.com/Order_Details.csv")
```

# Sales_order_header

```
sales_order_header.filter(col("Order ID") == 'B-25601').show()
```

# Sales_order_detail

```
sales_order_detail.filter(col("Order ID") == 'B-25601').show()
```

```
sales_order_detail.groupBy("Order Id").agg({'Order Id': 'count'}).filter(col("count(Order Id)") != 1).show()
```

# Header: count

```
sales_order_header.count() #560
sales_order_header.filter(col("Order id").isNotNull()).count() #500
```

# Header: count distinct

```
sales_order_header.select("Order Id").distinct().count() #501
```

# Header: null

```
sales_order_header.groupBy("Order Id").agg({'Order Id': 'count'}).filter(col("count(Order Id)") != 1).show()
```

# Header: check null

```
sales_order_header.filter(col("Order id").isNull()).show()
```
